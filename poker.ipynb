```python
class ShortDeck():
    def __init__(self):
        self.newdeck = self.__generate_deck()
    def __generate_deck(self):
        
        cards = []
        for suit_i in ['s','h','d','c']:
            for rank_i in ['6','7','8','9','T','J','Q','K','A']:
                cards.append((rank_i,suit_i))
        
        return cards
    def shuffle_cards(self):
        
        newdeck = self.newdeck.copy()
        random.shuffle(newdeck)
        
        self.shuffled_deck = newdeck
        
    def deal(self):
        try:
            return self.shuffled_deck.pop()
        except:
            return 'end of deck'
        
```

```python
def straight_check(cards_in):
    A_case = False
    ranks = []
    for card in cards_in:
        if card[0] == 'A':
            A_case = True
        ranks.append(value_lookup[card[0]])
    ranks = list(set(ranks))
    
    
    ranks.sort()
    ranks = [0] + ranks
    
    dif = np.cumsum(list(np.diff(ranks)-1))

    if Counter(dif).most_common()[0][1] >= 5 :
        out =  ranks[np.argwhere(dif == Counter(dif).most_common()[0][0])+1][-5:]
    else:
        out = []

    
    if A_case and len(out) == 0:
        ranks[-1] = 6
        ranks.sort()
        dif = np.cumsum(list(np.diff(ranks)-1))
        if Counter(dif).most_common()[0][1] >= 5 :
            out= [5,6,7,8,9]
        else:
            out= []
        
    
    return out
```

```python
def analyze_hand(cards):
  
    if len(cards) == 1:
        return(f'High Card {cards[0][0]}')
    elif len(cards) == 2:
        card1 = cards[0]
        card2 = cards[1]
        

        
        
        if card1[0] == card2[0]:
            return(f'Pair {card1[0]}')
        else:
            connector_res = connector_analyzer(card1,card2)
            suit_res = suit_analyzer(card1,card2)
            if (connector_res and suit_res):
                return('Suited Connector')
            elif connector_res:
                return('Connector')
            elif suit_res:
                return('Suited')
            else:
                return('Junk')
    elif len(cards) == 5:
        c1 =cards[0]
        c2 =cards[1]
        c3 =cards[2]
        c4 =cards[3]
        c5 =cards[4]
        
        connector_res = connector_analyzer(c1,c2,c3,c4,c5)
        suit_res = suit_analyzer(c1,c2,c3,c4,c5)
        face_ranks = []
        for card in cards:
            face_ranks.append(card[0])
        res = Counter(face_ranks)
        if connector_res and suit_res:
            return(('Straight Flush',50000))
        
        elif res.most_common()[0][1] == 4:
            return(('Four of a Kind',40000))
        elif suit_res:
            return(('Flush',30000))
        elif res.most_common()[0][1] == 3 and res.most_common()[1][1] == 2:
            return(('Full House',20000))
        elif res.most_common()[0][1] == 3:
            return(('Three of a kind',1000))
        elif connector_res:
            return(('Straight',10000))

        elif res.most_common()[0][1] == res.most_common()[1][1] ==2:
            return(('Two Pairs',200))
        elif res.most_common()[0][1] == 2:
            return(('Pair',100))
        elif res.most_common()[0][1] == 1:
            return(('High Card',10))
        else:
            return('Other')

    else:
        return ('Only 5,2,1 cards!')
```
